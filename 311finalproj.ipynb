```python
import pandas as pd
import numpy as np
```

```python
record = pd.read_csv('flight.csv')
record.head()
```

```python
len(record)
```

```python
record.Origin_City.nunique()
```

```python
record.Destination_City.nunique()
```

```python
cities = pd.DataFrame(record.Origin_City.unique(), columns=['city'])
cities['id'] = np.arange(0, len(cities))
cities = cities[['id', 'city']]
cities['id'] = cities['id'] + 1
cities['states'] = cities['city'].str.split(',').str[1]
cities['city'] = cities['city'].str.split(',').str[0]
cities.head()
```

```python
states = pd.DataFrame(cities.states.unique(), columns=['states'])
states['id'] = np.arange(0, len(states))
states = states[['id', 'states']]
states['id'] = states['id'] + 1
states.head()
```

```python
for index, row in cities.iterrows():
    s = row['states']
    states_id = states.loc[states['states'] == s, 'id'].values
    cities.at[index, 'states_id'] = states_id[0]
```

```python
cities.drop(columns=['states'], inplace=True)
cities['states_id'] = cities['states_id'].astype(int)
cities.head()
```

```python
cities.to_csv('city.csv', index=False)
```

```python
states.to_csv('state.csv', index=False)
```

```python
airlines = pd.DataFrame(record.Airline_Name.unique(), columns=['airline'])
airlines['id'] = np.arange(0, len(airlines))
airlines = airlines[['id', 'airline']]
airlines['id'] = airlines['id'] + 1
airlines.head()
```

```python
airlines.to_csv('airlines.csv', index=False)
```

```python
cancelled = pd.DataFrame(record.Cancellation_Code.unique(), columns=['cancellation_code'])
cancelled['id'] = np.arange(0, len(cancelled))
cancelled = cancelled[['id', 'cancellation_code']]
cancelled['id'] = cancelled['id'] + 1
cancelled.head()
```

```python
cancelled.to_csv('cancelled.csv', index=False)
```

```python
for index, row in record.iterrows():
    airline = row['Airline_Name']
    airline_id = airlines.loc[airlines['airline'] == airline, 'id'].values
    record.at[index, 'airline_id'] = airline_id[0]
```

```python
record.head()
```

```python
for index, row in record.iterrows():
    c = row['Origin_City']
    origin_city_id = cities.loc[cities['city'] == c, 'id'].values
    record.at[index, 'origin_city_id'] = origin_city_id[0]
```

```python
for index, row in record.iterrows():
    cancel = row['Cancellation_Code']
    cancellation_id = cancelled.loc[cancelled['cancellation_code'] == cancel, 'id'].values
    record.at[index, 'cancellation_id'] = cancellation_id[0]
```

```python
record.head()
```

```python
record = pd.merge(record, cities, left_on='Destination_City', right_on='city')
record.head()
```

```python
record.rename({'id': 'destination_city_id'}, axis=1, inplace=True)
record.head()
```

```python
flight_data = record[['Flight_Date', 'airline_id', 'origin_city_id', 'destination_city_id', 'Scheduled_Departure_Time', 
                      'Actual_Departure_Time', 'Departure_Delay_Minutes', 'Scheduled_Arrival_Time', 'Actual_Arrival_Time',
                      'Arrival_Delay_Minutes', 'cancellation_id']]

flight_data.head()
```

```python
flight_data[['airline_id', 'origin_city_id', 'cancellation_id']] = flight_data[['airline_id', 'origin_city_id', 'cancellation_id']].astype(int)
```

```python
flight_data['id'] = np.arange(0, len(flight_data))
flight_data['id'] = flight_data['id'] + 1
```

```python
flight_data.head()
```

```python
flight_data = flight_data[['id', 'Flight_Date', 'airline_id', 'origin_city_id', 'destination_city_id', 'Scheduled_Departure_Time', 
                      'Actual_Departure_Time', 'Departure_Delay_Minutes', 'Scheduled_Arrival_Time', 'Actual_Arrival_Time',
                      'Arrival_Delay_Minutes', 'cancellation_id']]
```

```python
flight_data.head()
```

```python
flight_data.to_csv('flight_data.csv', index=False)
```
